```python
import torch
```

```python
x = torch.tensor(3.5)
print("x=",x)

y = x + 3
print("y=",y)
```

```python
#PyTorch张量  requires_grad:自动求梯度的开关
x = torch.tensor(3.5,requires_grad=True)
# print(x)
y = (x-1) * (x-2) * (x-3)
# print(y)

# 计算梯度(一行代码完成)
# 梯度dy/dx = 3x^2 −12x + 11 
# x是3.5，所以梯度是3*（3.5*3.5）−12*（3.5）+ 11 = 5.75
y.backward()

#查看计算的梯度结果
x.grad
```
