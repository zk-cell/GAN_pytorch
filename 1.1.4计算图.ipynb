```python
# 当y以x的形式定义，而z以y的形式定义时，PyTorch便将这些张量连成一幅
# 图，以展示这些张量是如何连接的。这幅图叫计算图
import torch
```

```python


x = torch.tensor(3.5, requires_grad=True)
y = x**2
z = 2*y + 3
```

```python
# 计算梯度  z -> y -> x   dz/dx = 4x
z.backward()
```

```python
# x=3.5时的dz/dx
x.grad
# 注意:因为我们要求PyTorch使用z.backward ()从z反向计算。
# 因此，x.grad是dz/dx，而不是dy/dx。
```

```python
# 复杂的链式求导关系
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)
x = 2*a + 3*b
y = 5*a**2 + 3*b**3
z = 2*x + 3*y

# 计算梯度 
z.backward()
```

```python
# dz/da = 4 + 30a， a=2使用一行代码验证
a.grad
```
